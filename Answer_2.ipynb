```python
import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
import urllib
```

```python
url = 'https://www.imdb.com/list/ls091520106/'
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
movie_name = []
rating = []
year = []
```

```python
movie_data = soup.findAll('div', attrs= {'class': 'lister-item mode-detail'})
```

```python
movie_data
```

```python
 for store in movie_data:
    name = store.h3.a.text
    movie_name.append(name)
    
    year_of_release = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')','')
    year.append(year_of_release)
    
    rate = store.find('div', class_ = 'ipl-rating-star small').text.replace('\n','')
    rating.append(rate)
```

```python
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of release': year, 'Movie Rating': rating})
```
